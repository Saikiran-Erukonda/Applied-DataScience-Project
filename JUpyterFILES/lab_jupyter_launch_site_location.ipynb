```python
# TASK 3: Calculate the distances between a launch site to its proximities
import pandas as pd
import numpy as np

# Define Haversine function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in km
    lat1, lon1, lat2, lon2 = [np.radians(val) for val in [lat1, lon1, lat2, lon2]]  # Convert degrees to radians

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))

    return R * c  # Distance in km

# Sample DataFrame (Modify for actual proximity locations)
proximity_df = pd.DataFrame({
    'Name': ['Nearest Highway', 'Coastline', 'City Center'],
    'Lat': [28.5721, 28.600, 28.401],
    'Long': [-80.6480, -80.500, -80.700]
})

# Define launch site coordinates
launch_site_lat = spacex_df.iloc[0]['Lat']
launch_site_long = spacex_df.iloc[0]['Long']

# Compute distances to proximity locations
proximity_df['Distance (km)'] = proximity_df.apply(
    lambda row: haversine(launch_site_lat, launch_site_long, row['Lat'], row['Long']), axis=1
)

# Display results
proximity_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
